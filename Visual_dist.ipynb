### Visualizing Distributions

For this notebook i am going to use apple quality dataset for the visualization

```python
#importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading of the dataset
apple = pd.read_csv('C:\\Users\\user\\AppData\\Local\\Temp\\b3841061-ff9c-4bdf-9a91-f6db5145b9f4_archive (2).zip.9f4\\apple_quality.csv')
apple
```

```python
apple.isnull().sum()
```

```python
filtered_apple = apple.dropna(inplace = True)
filtered_apple
```

Plotting scatter plot to show the distribution of the apple dataset

```python
plt.title('Size Vs Sweetness')
plt.ylabel('Sweetness')
plt.xlabel('Size')
x = apple['Size']
y = apple['Sweetness']
plt.plot(x, y, 'bo')
plt.show()
```

Plotting histogram to display Weight of the apples

```python
apple['Weight'].plot(kind = 'hist', bins = 10, figsize = (12,6), color='green')
plt.show()
```

Plotting a 2D histogram to visualize Juiciness and Ripeness

```python
x = apple['Juiciness']
y = apple['Ripeness']
fig, ax = plt.subplots()
hist = ax.hist2d(x[::10], y[::10])
```

Plotting density plot to display the Crunchiness

```python
sns.kdeplot(apple['Crunchiness'])
plt.show()
```
