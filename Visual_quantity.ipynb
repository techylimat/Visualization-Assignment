### VISUALIZATION

```python
#Importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Loading of the dataset
birds = pd.read_csv('C:\\Users\\user\\Downloads\\birds.csv')
birds.head()
```

Visualizing MaxBodyMass using Line plot

```python
Bodymass = birds['MaxBodyMass'] 
Bodymass.plot()
```

Plotting scatter plot to visualize the category of birds with MinBodyMass

```python

filtered_data = birds[birds['MinLength'] <= 50]

# Scatter plot
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)
plt.scatter(filtered_data['Category'], filtered_data['MinBodyMass'], cmap='viridis', alpha=0.8, color='red')
plt.xlabel('Category')
plt.ylabel('MinBodyMass')
plt.title('Birds Distribution based on Category and MinBodyMass')
plt.show()
```

Plotting bar chart to visualize the Family of the birds

```python
Family_count = birds.value_counts(birds['Family'].values, sort=True)
plt.rcParams['figure.figsize'] = [6, 12]
Family_count.plot.barh()
```

Comparing the family of birds with minimum and maximum length

```python
minLength = birds['MinLength']
maxLength = birds['MaxLength']
family = birds['Family']

plt.barh(family, maxLength)
plt.barh(family, minLength)

plt.show()
```
